```python
import pandas as pd
cp= pd.read_excel("Survey Estimates.xlsx")
cp.head()
```

```python
null_cols=cp.columns[cp.isna().any()]
cp[null_cols].isna().sum()
```

```python
cp = cp.drop(["WAZ Survey Sample (N)", "HAZ Survey Sample (N)","Short Source", "Source", "SDG Region"], axis = 1)
cp.head()
```

```python
missing_values_count = cp.isnull().sum()
cp.head()
```

```python
cp = cp.dropna()
cp.head()
```

```python
# Fill the missing values with a mean
mean = cp.mean()
cp = cp.fillna(mean)
cp.head()
```

```python
# Identify and count the duplicate data
duplicate_rows_count = cp.duplicated().sum()

# Remove the duplicate data
cp = cp.drop_duplicates()
cp.head()
```

```python
cp.to_excel("Survey Estimates1.xlsx", index=False)
```
